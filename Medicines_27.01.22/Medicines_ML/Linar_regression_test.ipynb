### Import

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
```

### Data

```python
data = pd.read_csv('data_filter_score_full.csv', sep = ';')
data
```

```python
data_learning = pd.read_csv('NM+PROD_learn2.csv', sep = ';')
data_learning
```

```python
data_learning.corr()
```

```python
model = LinearRegression()
X = pd.DataFrame(np.c_[data_learning['Cosine_comp_NM'], data_learning['Number_comp_NM'],
                      data_learning['Word_eq_ru_NM'], data_learning['Every_word_mectric_NM'], 
                      data_learning['Levenshtein_metric_NM'],
                      data_learning['Cosine_comp_PROD'], data_learning['Word_eq_ru_PROD'], data_learning['Every_word_metric_PROD'],
                      data_learning['Levenshtein_metric_PROD']], columns=['Cosine_comp_NM','Number_comp_NM', 'Word_eq_ru_NM','Every_word_mectric_NM',
                                                                          'Levenshtein_metric_NM','Cosine_comp_PROD','Word_eq_ru_PROD','Every_word_metric_PROD',
                                                                          'Levenshtein_metric_PROD'])
Y = data_learning['MARK']

model.fit(X, Y)
pd.DataFrame(data = model.coef_, index = X.columns, columns = ['Coef.'])
```

```python
pred = pd.DataFrame(np.c_[data['Cosine_comp_NM'], data['Number_comp_NM'],
                      data['Word_eq_ru_NM'], data['Every_word_mectric_NM'], 
                      data['Levenshtein_metric_NM'],
                      data['Cosine_comp_PROD'], data['Word_eq_ru_PROD'], data['Every_word_metric_PROD'],
                      data['Levenshtein_metric_PROD']], columns=['Cosine_comp_NM','Number_comp_NM', 'Word_eq_ru_NM','Every_word_mectric_NM',
                                                                          'Levenshtein_metric_NM','Cosine_comp_PROD','Word_eq_ru_PROD','Every_word_metric_PROD',
                                                                          'Levenshtein_metric_PROD'])
out = model.predict(pred)
# for element in out:
#     with open('Linear_model_SCORE_2.txt', "a", encoding="utf-8") as file:
#         file.write(str(float(element)) + '\n')
```
